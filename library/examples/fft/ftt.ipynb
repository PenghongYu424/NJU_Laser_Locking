```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
import fft
```

```python
f_a_phi_list = [(1, 1, 0), (2, 0.5, 0), (3, 0.5, np.pi/4), (5, 0.5, np.pi/2), (6, 0.5, np.pi)]
offset = 0.1

x = np.linspace(0,10, 1000)
y = np.zeros(len(x))
y += offset
for f,a,phi in f_a_phi_list:
    y += a*np.cos(2*np.pi*f*x-phi)

fig = plt.figure(figsize=(5.2, 7.4))

ax1 = fig.add_subplot("311")
ax1.set_title("Input data")
ax1.set_ylabel("Y", fontsize=11)
ax1.set_xlabel("X", fontsize=11)
ax1.plot(x, y, linewidth=0.7, label="input data")
ax1.set_ylim([-3,3])
#ax1.legend(loc="best")

ax1 = fig.add_subplot("312")
f,A = fft.fft_fixed_rate(y, period=x[1]-x[0], return_complex=True)
ax1.set_title("FFT data")
ax1.set_ylabel("Amplitude A", fontsize=11)
ax1.set_xlabel("Frequency", fontsize=11)
ax1.plot(f, np.real(A), linewidth=0.7, label="Re(A)")
ax1.plot(f, np.imag(A), linewidth=0.7, label="Im(A)")
ax1.set_ylim([-1.1,1.1])
ax1.legend(loc="best")

ax1 = fig.add_subplot("313")
f,A = fft.fft_fixed_rate(y, period=x[1]-x[0], return_complex=False)
ax1.set_title("FFT data (power spectrum)")
ax1.set_ylabel("Amplitude A", fontsize=11)
ax1.set_xlabel("Frequency", fontsize=11)
ax1.plot(f, A, linewidth=0.7, label="Abs(A)")
ax1.set_ylim([-0.1,1.1])
ax1.legend(loc="best")

plt.tight_layout()
plt.show()
```
