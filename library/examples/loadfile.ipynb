```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
from files.picoscope import load_file, convert_unit_from_SI, convert_unit_to_SI
```

```python
convert_unit_from_SI(convert_unit_to_SI(0.1234, "uV"), "SI")
```

```python

fig = plt.figure(figsize=(5.2, 9.4))


x, y = load_file("./testdata/test.txt", ("F", "MHz"), ("A", "dBV"))

ax1 = fig.add_subplot("411")
ax1.set_title("Sensor spectrum")
ax1.set_ylabel("Amplitude (dBV)", fontsize=11)
ax1.set_xlabel("Frequency (MHz)", fontsize=11)
#ax1.set_yscale("log")
ax1.plot(x, y, linewidth=0.7, label="dBV/MHz")
ax1.legend(loc="best")


x, y = load_file("./testdata/test.txt", ("F", "MHz"), ("A", "V"))

ax1 = fig.add_subplot("412")
ax1.set_title("Sensor spectrum")
ax1.set_ylabel("Amplitude (V)", fontsize=11)
ax1.set_xlabel("Frequency (MHz)", fontsize=11)
#ax1.set_yscale("log")
ax1.plot(x, y, linewidth=0.7, label="V/MHz")
ax1.legend(loc="best")


x, y = load_file("./testdata/test.txt", "F", ("A", "dBu"))

ax1 = fig.add_subplot("413")
ax1.set_title("Sensor spectrum")
ax1.set_ylabel("Amplitude (V)", fontsize=11)
ax1.set_xlabel("Frequency (Hz)", fontsize=11)
#ax1.set_yscale("log")
ax1.plot(x, y, linewidth=0.7, label="SI/SI")
ax1.legend(loc="best")


x, y = load_file("./testdata/test.txt", ("F", None), ("A", None))

ax1 = fig.add_subplot("414")
ax1.set_title("Sensor spectrum")
ax1.set_ylabel("Amplitude (raw)", fontsize=11)
ax1.set_xlabel("Frequency (raw)", fontsize=11)
#ax1.set_yscale("log")
ax1.plot(x, y, linewidth=0.7, label="raw/raw")
ax1.legend(loc="best")

plt.tight_layout()
plt.show()
```
